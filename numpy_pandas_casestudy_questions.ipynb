## ðŸ§  **Question 1: NumPy Array Manipulation**
1. Create a 1D NumPy array with integers from **1 to 24**.  
2. Reshape the array into a **4x6 matrix**.  
3. Extract the **third row** and **second column** from the reshaped array.  
4. Transpose the matrix and display the result.

```python
# Solution for Question 1
import numpy as np

# 1. Create 1D array from 1 to 24
arr = np.arange(1, 25)
print("1D array:")
print(arr)
print("\n")

# 2. Reshape into 4x6 matrix
matrix = arr.reshape(4, 6)
print("4x6 matrix:")
print(matrix)
print("\n")

# 3. Extract third row and second column
third_row = matrix[2]  # remember: zero-based indexing
second_col = matrix[:, 1]
print("Third row:")
print(third_row)
print("\nSecond column:")
print(second_col)
print("\n")

# 4. Transpose the matrix
transposed = matrix.T
print("Transposed matrix:")
print(transposed)
```

## ðŸ§  **Question 2: Array Element Extraction**
1. Create a **5x5 NumPy array** with random integers between **10 and 100**.  
2. Extract all the **even numbers** from the array.  
3. Replace all numbers **greater than 50** with **0**.  
4. Extract the diagonal elements of the array.

```python
# Solution for Question 2
import numpy as np

# 1. Create 5x5 array with random integers between 10 and 100
arr = np.random.randint(10, 101, size=(5, 5))
print("Original 5x5 array:")
print(arr)
print("\n")

# 2. Extract even numbers
even_numbers = arr[arr % 2 == 0]
print("Even numbers from the array:")
print(even_numbers)
print("\n")

# 3. Replace numbers > 50 with 0
arr_modified = np.where(arr > 50, 0, arr)
print("Array with numbers > 50 replaced by 0:")
print(arr_modified)
print("\n")

# 4. Extract diagonal elements
diagonal = np.diag(arr)
print("Diagonal elements:")
print(diagonal)
```

## ðŸ§  **Question 3: DataFrame Creation and Operations (Pandas)**
1. Create a Pandas DataFrame containing the following data:

| Student_ID | Name   | Subject | Marks |
|------------|--------|---------|-------|
| 1          | Alice  | Math    | 88    |
| 2          | Bob    | Science | 75    |
| 3          | Carol  | History | 92    |
| 4          | David  | Math    | 85    |
| 5          | Eva    | Science | 90    |

2. Find the **average marks** for each subject.  
3. Add a new column named **Grade** based on the marks:
- Marks â‰¥ 90 â†’ A
- Marks â‰¥ 80 â†’ B
- Marks â‰¥ 70 â†’ C
- Else â†’ D

```python
# Solution for Question 3
import pandas as pd

# 1. Create the DataFrame
data = {
    'Student_ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eva'],
    'Subject': ['Math', 'Science', 'History', 'Math', 'Science'],
    'Marks': [88, 75, 92, 85, 90]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print("\n")

# 2. Calculate average marks for each subject
subject_averages = df.groupby('Subject')['Marks'].mean()
print("Average marks by subject:")
print(subject_averages)
print("\n")

# 3. Add Grade column
def assign_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    elif marks >= 70:
        return 'C'
    else:
        return 'D'

df['Grade'] = df['Marks'].apply(assign_grade)
print("DataFrame with Grades:")
print(df)
```

## ðŸ§  **Question 4: Dataset Analysis (Pandas, Matplotlib, Seaborn)**
Use the dataset `diabetes.csv` and perform the following tasks:

1. Display the **BMI** and **Blood Pressure** for samples with **index ranging from 20 to 40**.  
2. Find how many patients have **BMI > 30**.  
3. Find the **mean and standard deviation** of the **Glucose** levels.  
4. Plot a **boxplot** for the column **Insulin**.  
5. Create a **heatmap** showing the correlation between the numerical columns in the dataset.

```python
# Solution for Question 4
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('C:/ICTAK DSA/Case study #4/diabetes_model.csv')

# 1. Display BMI and Blood Pressure for samples with index 20-40
print("\n1. BMI and Blood Pressure (index 20-40):")
print(df.loc[20:40, ['BMI', 'BloodPressure']])

# 2. Count patients with BMI > 30
bmi_count = len(df[df['BMI'] > 30])
print(f"\n2. Number of patients with BMI > 30: {bmi_count}")

# 3. Mean and standard deviation of Glucose levels
glucose_mean = df['Glucose'].mean()
glucose_std = df['Glucose'].std()
print(f"\n3. Glucose Statistics:")
print(f"Mean: {glucose_mean:.2f}")
print(f"Standard Deviation: {glucose_std:.2f}")

# 4. Create boxplot for Insulin
print("\n4. Boxplot for Insulin:")
plt.figure(figsize=(10, 6))
plt.title('Boxplot of Insulin Levels')
sns.boxplot(x=df['Insulin'])
plt.show()

# 5. Create correlation heatmap
print("\n5. Heatmap showing the correlation between the numerical columns in the dataset:")
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
